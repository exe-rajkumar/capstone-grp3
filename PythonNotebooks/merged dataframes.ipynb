```python
import pandas as pd
import numpy as np
import os
import ast
```

```python
cast=pd.read_csv(r'C:\Users\sambangi.bindu\Downloads\cast.csv')
crew=pd.read_csv(r'C:\Users\sambangi.bindu\Downloads\crew.csv')
collections=pd.read_csv(r'C:\Users\sambangi.bindu\Downloads\collection.csv')
genre=pd.read_csv(r'C:\Users\sambangi.bindu\Downloads\genres.csv')
prodcompanies=pd.read_csv(r'C:\Users\sambangi.bindu\Downloads\Production_companies.csv')
prodcountries=pd.read_csv(r'C:\Users\sambangi.bindu\Downloads\Production_countries.csv')
spokenlang=pd.read_csv(r'C:\Users\sambangi.bindu\Downloads\spoken_languages.csv')
moviedata=pd.read_csv(r'C:\Users\sambangi.bindu\Downloads\metadata.csv')
rating=pd.read_csv(r'C:\Users\sambangi.bindu\Downloads\rating.csv')
```

```python
a=moviedata[(moviedata['status']!='Rumored') & (moviedata['popularity']<=25)]
a.drop(columns=['original_title'],inplace=True)
a
```

```python
years_to_filter = [2013]
a['release_date'] = pd.to_datetime(a['release_date'])
a['year'] = a['release_date'].dt.year
filtered_df = a[a['year'].isin(years_to_filter)]
filtered_df.drop(columns=['year'], inplace=True)
filtered_df
```

```python
months_to_filter = [1,2,3,4,5,6,7,8,9]
filtered_df['release_date'] = pd.to_datetime(filtered_df['release_date'])
filtered_df['month'] = filtered_df['release_date'].dt.month
filtered_data = filtered_df[filtered_df['month'].isin(months_to_filter)]
filtered_data
```

```python
filtered_df['month'].unique()
```

```python
filtered_df.to_csv('sample.csv',index=False)
```

```python
b=filtered_data.merge(genre,on='mov_id',how='left')
b.drop(columns=['adult','video','vote_count'],inplace=True)
b
```

```python
c=b.merge(rating,left_on='mov_id',right_on='tmdbId',how='left')
c.drop(columns=['tmdbId'],inplace=True)
c
```

```python
d=c.merge(cast,on='mov_id',how='left')
d.drop(columns=['credit_id','order','profile_path'],inplace=True)
d.rename(columns={'gender':'cast_geneder','name':'cast_name','id':'cast_id'},inplace=True)
d
```

```python
e=d.merge(crew,on='mov_id',how='left')
e.drop(columns=['Unnamed: 0','credit_id','profile_path'],inplace=True)
e.rename(columns={'department':'crew_department','gender':'crew_gender','job':'crew_job','name':'crew_name','id':'crew_id'},inplace=True)
e
```

```python
f=e.merge(prodcompanies,left_on='mov_id',right_on='movie_id',how='left')
f.drop(columns=['movie_id'],inplace=True)
f.rename(columns={'name':'prodcompany_name','id':'prodcompany_id'},inplace=True)
f
```

```python
g=f.merge(prodcountries,left_on='mov_id',right_on='movie_id',how='left')
g.drop(columns=['movie_id'],inplace=True)
g.rename(columns={'iso_3166_1':'prodcountry_code','name':'prodcountry_name'},inplace=True)
g
```
