```python
meta['id'] = pd.to_numeric(meta['id'], errors='coerce')
meta = meta.dropna(subset=['id'], how='any')
```

# Genres dataframe

```python
import copy
tempcont=copy.deepcopy(meta[meta['genres'].isnull()==False])
tempcont
```

```python
tempcont['genres'] = tempcont['genres'].apply(ast.literal_eval)
```

```python
for index,l1 in tempcont.iterrows():
    movid=l1['id']
    #print(movid)
    for dic in l1['genres']:
        dic['mov_id']=movid
```

```python
a=[]
for i in tempcont['genres']:
    for dic in i:
        a.append(dic)
```

```python
genres=pd.DataFrame(a)
```
