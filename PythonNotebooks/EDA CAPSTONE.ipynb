```python
plt.figure(figsize=(8, 6))
plt.hist(df['popularity'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Popularity')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(df, y='revenue')
plt.ylabel('Revenue')
plt.title('Box Plot of Revenue')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['vote_average'], df['popularity'])
plt.xlabel('Vote Average')
plt.ylabel('Popularity')
plt.title('Scatter Plot: Vote Average vs. Popularity')
plt.show()
#to find whether vote avg and popularity have any relation or not
```

```python
a=df.groupby('original_language')['revenue'].mean().nlargest(20).sort_values(ascending=False)
a
```

```python
a.plot(kind='bar')
plt.xticks(rotation=45)
plt.title('Top languages that are producing more revenue')
```

```python
plt.figure(figsize=(10, 6))
df['prodcountry_name'].value_counts().plot(kind='bar', color='purple')
plt.xlabel('Production Country')
plt.ylabel('Number of Movies')
plt.title('Production Country Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
df['prodcountry_name'].value_counts().nlargest(10).plot(kind='bar')
plt.xticks(rotation=45,ha='right')
plt.title('Production Country distribution')
```
